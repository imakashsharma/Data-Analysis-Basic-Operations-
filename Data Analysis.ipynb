```python
#simple feature scalling method
import pandas as pd
import numpy as np
csv_path='info.csv' 
df=pd.read_csv(csv_path)
df["length"] = df["length"]/df["length"].max()
df.head(6)
```

```python
#min-max method
import pandas as pd
import numpy as np
csv_path='info.csv'
df=pd.read_csv(csv_path)
df["length"] = (df["length"]-df["length"].min())-(df["length"].max()-df["length"].min())
df.head(6)
```

```python
#Z-Score method
import pandas as pd
import numpy as np
csv_path='info.csv'
df=pd.read_csv(csv_path)
df["length"] = (df["length"]-df["length"].mean())/df["length"].std()
df.head(6)
```

```python
#binning
import pandas as pd
import numpy as np
csv_path='price.csv'
df=pd.read_csv(csv_path)
bins=np.linspace(min(df["price"]),max(df["price"]),4)
group_names=["Low","Medium","High"]
df["price-binned"] = pd.cut(df["price"],bins,labels=group_names,include_lowest=True)
df.head(20)
```

```python
#linear regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pd.read_csv('marks.csv')
df
```

```python
%matplotlib inline
plt.xlabel('maths')
plt.ylabel('english')
plt.scatter(df.maths,df.english, color='red' ,marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['maths']],df.english)
```

```python
reg.predict([[5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=m*x+b
-0.37846551*5+7.559638942617665
#y=coef
#x=predicting value
#y=intercept
```

```python
%matplotlib inline
plt.xlabel('English Marks', fontsize=15)
plt.ylabel('Maths Marks', fontsize=15)
plt.scatter(df.english,df.maths, color='red' ,marker='*')
plt.plot(df.english,reg.predict(df[['english']]),color='blue')
```

```python
#multi varient regression model
import pandas as pd
import numpy as np
from sklearn import linear_model
df=pd.read_csv('house.csv')
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,5]])
```

```python
#price=m1(coefficient 1)*area + m2(coefficient 2)*bedrooms + me(coefficient 3)*area + b(intercept)
5.99099799e+02*3000+-3.44639919e+05*3+-3.19162330e+02*5+80339.09686083614
```
